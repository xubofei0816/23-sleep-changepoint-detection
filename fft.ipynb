```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample time history
csv_file_path = 'data/asleep/08db4255286f_index1_awake0.csv'

df = pd.read_csv(csv_file_path)

# Extract the acceleration data from the specified column
enmo_column = 'enmo'
anglez_column = 'anglez'
acceleration_history = df[enmo_column]

# Determine the sampling frequency and number of data points
fs = 0.2  # 0.2 samples per second (0.2 Hz)
N = len(acceleration_history)

# Compute the FFT
fft_result = np.fft.fft(acceleration_history)

# Compute the corresponding frequencies
frequencies = np.fft.fftfreq(N, 1.0 / fs)

# Calculate the magnitude of the FFT result (absolute value)
magnitude = np.abs(fft_result)

# Create a plot to visualize the frequency content
plt.figure(figsize=(10, 6))
plt.semilogx(frequencies[:N // 2], magnitude[:N // 2])  # Plot only the positive frequencies
plt.title("Frequency Content of Acceleration Time History")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()

```

```python
#fft of enmo data - asleep
import numpy as np
import matplotlib.pyplot as plt
import os

# Directory containing the CSV files
data_dir = 'data/asleep'

series_id = "08db4255286f"

# List all CSV files in the directory
csv_files = [file for file in os.listdir(data_dir) if file.endswith('.csv') and series_id in file]

# Initialize lists to store frequency and magnitude data
all_frequencies = []
all_magnitudes = []

# Initialize a list to store file names without extensions
file_labels = []

# Loop through the CSV files
for csv_file in csv_files:
    # Construct the full file path
    file_path = os.path.join(data_dir, csv_file)

    # Read the CSV file into a pandas DataFrame (assuming it has a 'Acceleration' column)
    df = pd.read_csv(file_path)
    enmo_column = 'enmo'
    anglez_column = 'anglez'
    acceleration_history = df[enmo_column]

    # Determine the sampling frequency and number of data points
    fs = 0.2  # 0.2 samples per second (0.2 Hz)
    N = len(acceleration_history)

    # Compute the FFT
    fft_result = np.fft.fft(acceleration_history)

    # Compute the corresponding frequencies
    frequencies = np.fft.fftfreq(N, 1.0 / fs)

    # Calculate the magnitude of the FFT result (absolute value)
    magnitude = np.abs(fft_result)

    # Extract the file name without extension
    file_name = os.path.splitext(csv_file)[0]
    file_labels.append(file_name)

    # Append the frequencies and magnitudes to the lists
    all_frequencies.append(frequencies[:N // 2])
    all_magnitudes.append(magnitude[:N // 2])

f_low = 0.00001

# Create a plot to visualize the frequency content of all files on the same graph
plt.figure(figsize=(10, 6))
for i in range(len(csv_files)):
#for i in range(4,5):
    # Filter the data to the desired frequency range
    mask = (all_frequencies[i] >= f_low)# & (all_frequencies[i] <= f_high)
    filtered_frequencies = all_frequencies[i][mask]
    filtered_magnitudes = all_magnitudes[i][mask]

    # Append the filtered data to the lists
    all_frequencies[i] = filtered_frequencies
    all_magnitudes[i] = filtered_magnitudes
    plt.loglog(all_frequencies[i], all_magnitudes[i], label=file_labels[i])

plt.title("Frequency Content of Acceleration Data from asleep segments (enmo)")
plt.xlabel("Frequency (Hz) - Log Scale")
plt.ylabel("Magnitude")
#plt.legend()
plt.grid(True)
plt.show()
```
